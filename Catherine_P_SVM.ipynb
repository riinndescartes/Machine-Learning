````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
````

````python
# Muat data
data = pd.read_csv('bank-additional-full.csv', sep=';')

# Tampilkan 5 data pertama
print(data.head())
````

````python
data.info()
````

````python
# Pisahkan data fitur dan target
X = data.drop('y', axis=1)
y = data['y']

# Ubah target menjadi 0 dan 1 (yes/no)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Ubah fitur kategorikal menjadi numerik
X = pd.get_dummies(X, drop_first=True)
````

````python
# Bagi data menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalisasi fitur
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
# Buat dan latih model SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
````

````python
# Prediksi data uji
y_pred = svm_model.predict(X_test)

# Evaluasi hasil
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
````

**1.1. Confusion Matrix:**

Kita mengukur kinerja model SVM yang menghasilkan *binary classification* menggunakan *confusion matrix*. Matriks ini memetakan kinerja model saat memproses data prediksi dibandingkan kenyataannya.

  [[10725    243]
                      
   [  950      439]]

Merupakan nilai dari Confusion Matrix berikut,

[[TN    FP]
                      
 [  FN     TP]]

**Keterangan Matriks:**
- TP = TRUE POSITIVE, model sukses memprediksi positif ("ya"), karena kenyataan memang positif ("ya").
- TN = TRUE NEGAITVE, model sukses memprediksi negative ("tidak"), karena kenyataan memang negative ("tidak").
- FP = FALSE POSITIVE, model sukses memprediksi positif ("ya"), namun salah karena kenyataannya negatif ("tidak").
- FN = FALSE NEGATIVE, model sukses memprediksi negative ("ya"), karena kenyataan memang negative ("ya").

Maka Confusion Matrix memberikan informasi bahwa, dari nilai TP dan TN artinya model dapat memprediksi dengan benar keadaan yang seharusnya 'ya' dan 'tidak' sebanyak 10.725 dan 439 kasus. Sedangkan dari nilai FP dan FN artinya model tidak dapat memprediksi dengan benar (gagal prediksi) keadaan yang seharusnya 'ya' dan 'tidak' sebanyak 243 dan 950 kasus.

Kita juga mendapatkan insight bahwa data berlabel 0 atau 'tidak' jumlahnya lebih banyak daripada data berlabel 1 atau 'ya'.



Kinerja model bisa disebut ideal ketika FN dan FP bernilai 0 terutama misalnya untuk dataset yang rentan terhadap FN dan FP, anggap saja data perbankan atau saham. Tetapi dari sumber literatur yang saya baca keadaan ideal ini jarang terwujud apalagi ketika berhadapan dengan dataset di dunia industri.


Pada perhitungan manual sebenarnya Confusion Matrix bisa digunakan untuk perhitungan akurasi model.  Contohnya seperti ini:

- `accuracy = ((TP+TN) / (TP+TN+FP+FN))`

Kita coba buktikan nilai akurasi dengan cara perhitungan manual.

             = 11.164/12.357
             = 0,90345553127781824067330258153273
             = 0,90

a. Precision dan Recall dalam Konteks Bisnis atau Domain

- Kelas 0 (Mayoritas): Karena kelas 0 memiliki precision yang tinggi (0.92), ini menunjukkan bahwa model sangat berhati-hati dalam memastikan data kelas 0 tidak salah diprediksi sebagai kelas 1. Di beberapa aplikasi, seperti pada kasus dengan risiko finansial rendah atau moderat, ini mungkin cukup memadai.

- Kelas 1 (Minoritas): Recall rendah (0.32) pada kelas 1 menunjukkan model tidak dapat menangkap sebagian besar data yang sebenarnya termasuk dalam kelas 1. Jika kelas 1 berhubungan dengan kejadian berisiko tinggi (seperti penipuan atau deteksi penyakit serius), recall rendah ini berarti banyak kasus kritis yang terlewat, berpotensi merugikan. Precision yang rendah di kelas 1 juga menunjukkan banyak data yang sebenarnya kelas 0 justru dianggap sebagai kelas 1, yang bisa berarti peningkatan jumlah false alarms.

b. Implikasi dari Macro Average dan Weighted Average

- Macro Average (Precision: 0.78, Recall: 0.65, F1-Score: 0.69): Rata-rata ini menghitung rata-rata dari semua kelas tanpa memperhitungkan jumlah data di setiap kelas, memberi pandangan lebih netral tentang performa keseluruhan. Nilai precision dan recall yang moderat (0.78 dan 0.65) di macro average ini mengindikasikan bahwa model belum optimal dalam menangani kelas minoritas dan menunjukkan kinerja yang bias ke kelas mayoritas.

- Weighted Average (Precision: 0.89, Recall: 0.90, F1-Score: 0.89): Rata-rata tertimbang ini, di sisi lain, menunjukkan hasil yang lebih tinggi karena didominasi oleh performa di kelas 0. Karena kelas 0 jauh lebih banyak jumlahnya, weighted average ini mungkin tampak memberikan performa yang baik, tetapi perlu diingat bahwa kelas 1 tetap bermasalah.


c. Analisis Kesalahan (Error Analysis)

- False Positives (FP) dan False Negatives (FN): Dari matriks konfusi, kita tahu terdapat 243 kesalahan False Positives untuk kelas 0 dan 950 kesalahan False Negatives untuk kelas 1. Ini menunjukkan bahwa model lebih sering gagal dalam mengidentifikasi kelas 1 (yang mungkin krusial) dibandingkan kesalahan pada kelas 0. FP dan FN yang tinggi di kelas 1 bisa berimplikasi signifikan di aplikasi real-world. Misalnya, jika model diterapkan di sektor keuangan untuk mendeteksi anomali, maka 950 FN artinya ada 950 kejadian penting yang tidak terdeteksi, sementara 243 FP bisa menyebabkan alarm palsu yang menguras sumber daya untuk diperiksa.


d. Evaluasi Terhadap Akurasi
- Meskipun akurasi model mencapai 90%, hal ini dapat menyesatkan karena mayoritas prediksi yang benar adalah untuk kelas 0. Pada masalah ketidakseimbangan kelas, akurasi menjadi kurang informatif sebagai metrik kinerja. Jika tujuan utama analisis adalah untuk mendeteksi kelas minoritas (kelas 1), maka akurasi tinggi tidak serta merta menjamin model yang baik.
Untuk itu, metrik yang lebih fokus pada performa di kelas 1 (seperti F1-score kelas 1, atau recall dan precision per kelas) akan lebih relevan untuk dijadikan dasar keputusan.

###a. Kernel Linear Undersampling


Undersampling dataset untuk mempersingkat proses training SVM.

````python
#Mendefinisikan ulang x dan y setelah undersampling.
x = data.drop('y', axis=1)
y = data['y']
````

````python
from imblearn.under_sampling import RandomUnderSampler

# c. Buat objek RandomUnderSampler
undersampler = RandomUnderSampler(random_state=42, sampling_strategy= "auto")

# d. Lakukan undersampling pada data
x_resampled, y_resampled = undersampler.fit_resample(x, y)
````

````python
# e. Ubah target menjadi 0 dan 1 (yes/no)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_resampled)

# f. Ubah fitur kategorikal menjadi numerik
X = pd.get_dummies(x_resampled, drop_first=True)
````

````python
# Bagi data menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalisasi fitur
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
# Buat dan latih model SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
````

````python
# Prediksi data uji
y_pred = svm_model.predict(X_test)

# Evaluasi hasil
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
````

#6. Mengapa Kernel Tertentu Mungkin Lebih Cocok

- Kernel Polynomial berfungsi baik dalam beberapa situasi, terutama ketika hubungan antara fitur dan kelas tidak linear, tetapi pemilihan derajat yang tepat sangat penting. Pada derajat yang lebih tinggi, model bisa menjadi terlalu kompleks dan menyebabkan overfitting, seperti yang terlihat pada degree 4 dan 5.

- Kernel RBF cenderung lebih fleksibel dan sering kali lebih baik dalam menangkap pola kompleks dalam data. Dengan pengaturan parameter C yang tepat, model dapat mengontrol margin dan kesalahan klasifikasi, sehingga dapat menghasilkan performa yang lebih baik dalam banyak kasus.

- Dalam dataset ini, kernel RBF dengan C = 1 memberikan hasil yang lebih konsisten dan lebih tinggi pada F1-Score, menunjukkan bahwa ia lebih mampu menangkap kompleksitas dalam data dibandingkan dengan kernel Polynomial pada derajat yang lebih tinggi.
